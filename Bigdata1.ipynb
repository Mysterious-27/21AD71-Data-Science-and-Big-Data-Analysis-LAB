```python
import matplotlib.pyplot as plt

hrs = [1, 2, 3, 4, 5, 6, 7, 8, 9]
ex = [88, 89, 79, 70, 80, 81, 89, 90, 99]

plt.plot(hrs, ex, linestyle='-', marker='*', color='red')
plt.xlabel('Hours')
plt.ylabel('Exam Score')
plt.grid(True)
plt.title('Hours vs Exam Score')
plt.show()
```
